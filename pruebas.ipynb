```python
import sqlite3
import time
import zlib

conn = sqlite3.connect('covid.sqlite')
cur = conn.cursor()

cur.execute('SELECT fecha, total_cases, new_cases FROM dias')
cases = dict()
for filas in cur:
    cases[filas[0]] = (filas[1], filas[2])
```

```python
dates = list()
for date in cases.keys():
    dates.append(date)
```
